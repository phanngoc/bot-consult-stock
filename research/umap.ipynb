```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import umap

# Create a synthetic dataset (3 blobs)
X, y = make_blobs(n_samples=300, centers=3, random_state=42)
```

```python
print("X shape:", X.shape, y.shape)
print("X:", X[:5])
```

```python

# Create a UMAP model
umap_model = umap.UMAP(n_neighbors=15, n_components=2, min_dist=0.1)

# Fit and transform the data using UMAP
embedding = umap_model.fit_transform(X)
```

```python

# Plot the original dataset
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title("Original Data")

# Plot the UMAP visualization
plt.subplot(122)
plt.scatter(embedding[:, 0], embedding[:, 1], c=y, cmap='viridis')
plt.title("UMAP Visualization")

plt.show()
```
