```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = ['This framework generates embeddings for each input sentence',
    'Sai lầm chiến thuật có thể làm suy yếu phòng tuyến Nga ở Ukraine', 
    'The quick brown fox jumps over the lazy dog.']
sentence_embeddings = model.encode(sentences)
```

```python
for sentence, embedding in zip(sentences, sentence_embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
```

```python
from importlib import reload
import gensim
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
from gensim.models.word2vec import LineSentence
import multiprocessing
import pandas as pd
cores = multiprocessing.cpu_count()
print(cores)

import regress_adapter
reload(regress_adapter)

data = regress_adapter.get_data('TPB', '2022-01-01', '2023-09-29')
print(data.head())
```

```python
# use undersea to tokenize data_list to array sentence
from underthesea import sent_tokenize

def token_sent(t):
    a = sent_tokenize(t)
    a = [str(x.strip()) for x in a]
    return a

data_list = data['text'].values.tolist()
print(token_sent(data_list[0]))
```

```python
arr_embedding = []

for t in data_list[:100]:
    sentences = token_sent(t)
    sentence_embeddings = model.encode(sentences)
    arr_embedding.append(sentence_embeddings)
```

```python
import numpy as np
print(len(arr_embedding))
print('arr_embedding:0:', len(arr_embedding[0]), len(arr_embedding[0][0]))
print('arr_embedding:1:', len(arr_embedding[1]), len(token_sent(data_list[1])))
print('arr_embedding:50:', len(arr_embedding[50]), len(token_sent(data_list[50])))
```
